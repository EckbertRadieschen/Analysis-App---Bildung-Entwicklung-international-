```python
import pandas as pd
import json
import re
from utils.hilfsfunktionen import short_analysis_frame

df_dev_data = pd.read_csv(r"data\raw\WDICSV.csv", encoding="utf-8")

indicator_list = list(df_dev_data["Indicator Name"].unique())

indicators_a_h = []
indicators_i_p = []
indicators_q_z = []
other_indicators = []

for indicator in indicator_list:
    if re.match(r"^[a-h]", indicator.lower()):
        indicators_a_h.append(indicator)
    elif re.match(r"^[i-p]", indicator.lower()):
        indicators_i_p.append(indicator)
    elif re.match(r"^[q-z]", indicator.lower()):
        indicators_q_z.append(indicator)
    else:
        other_indicators.append(indicator)

indicators_a_h
```

```python
variable_list = [
    'GDP per capita (constant 2015 US$)', 
    'Inflation, consumer prices (annual %)', 
    'Unemployment, total (% of total labor force) (modeled ILO estimate)', 
    'Labor force with advanced education (% of total working-age population with advanced education)', 
    'Manufacturing, value added (% of GDP)', 
    'Industry (including construction), value added (% of GDP)',  
    'Research and development expenditure (% of GDP)', 
    'Exports of goods and services (% of GDP)', 
    'Imports of goods and services (% of GDP)', 
    'High-technology exports (% of manufactured exports)',
    'Life expectancy at birth, total (years)', 
    'Infant', 
    'Mortality rate, under-5 (per 1,000 live births)', 
    'Maternal mortality ratio (modeled estimate, per 100,000 live births)', 
    'Physicians (per 1,000 people)', 
    'Prevalence of HIV, total (% of population ages 15-49)', 
    'Individuals using the Internet (% of population)', 
    'Renewable energy consumption (% of total final energy consumption)', 
    'Total greenhouse gas emissions excluding LULUCF per capita (t CO2e/capita)', 
    'CO2 emissions', 
    'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)', 
    'Poverty headcount ratio at $8.30 a day (2021 PPP) (% of population)', 
    'Gini index', 
    'Intentional homicides (per 100,000 people)', 
    'Political Stability - Governance estimate (approx. -2.5 to +2.5)', 
    'Government Effectiveness - Governance estimate (approx. -2.5 to +2.5)', 
    'New business density (new registrations per 1,000 people ages 15-64)', 
    'Trademark applications, resident, by count'
]

df_dev = pd.read_csv(r"data\raw\WDICSV.csv", encoding="utf-8")

variable_list_df = list(df_dev["Indicator Name"].unique())

relevant_variables = []
found_variables = []

for var_df in variable_list_df:
    for var_list in variable_list:
        if var_list.lower() in var_df.lower():
            relevant_variables.append(var_df)
            if not var_list in found_variables:
                found_variables.append(var_list)
            break


missed_variables_list = []

for var in variable_list:
    if not var in found_variables:
        missed_variables_list.append(var)

relevant_variables
```
